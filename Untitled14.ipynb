```python
import matplotlib.pyplot as plt

input_sizes = [5, 10, 20, 100, 1000, 5000, 10000]
insertion_sort_times = [0, 0, 0, 2, 1, 4, 12]
selection_sort_times = [0, 0, 0, 1, 1, 6, 44]
bubble_sort_times = [0, 0, 0, 0, 1, 30, 130]

plt.figure(figsize=(8, 3))
plt.plot(input_sizes, insertion_sort_times, marker='s', label='Insertion Sort')
plt.plot(input_sizes, selection_sort_times, marker='^', label='Selection Sort')
plt.plot(input_sizes, bubble_sort_times, marker='o', label='Bubble Sort')
plt.title('Sorting Algorithm Benchmark', fontsize=16)
plt.xlabel('Input Size (n)', fontsize=10)
plt.ylabel('Runtime (ms)', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.9)
plt.legend(fontsize=14)
plt.tight_layout()
plt.show()
```
