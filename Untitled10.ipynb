```python
import time
import numpy as np
import platform
import psutil
import matplotlib.pyplot as plt
from cpuinfo import get_cpu_info


def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


def benchmark_sorting_algorithms():
    sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000]
    algorithms = [insertion_sort, selection_sort, bubble_sort]
    results = {alg.__name__: [] for alg in algorithms}

    for size in sizes:
        for alg in algorithms:
            arr_copy = np.random.randint(0, 10000, size).tolist()
            start_time = time.time()
            alg(arr_copy)
            end_time = time.time()
            results[alg.__name__].append(end_time - start_time)


    plt.figure(figsize=(10, 6))
    for alg_name, runtimes in results.items():
        plt.plot(sizes, runtimes, label=alg_name, marker="o")

    plt.xlabel("Input Size (n)")
    plt.ylabel("Runtime (seconds)")
    plt.title("Sorting Algorithm Benchmark")
    plt.legend()
    plt.yscale("log")
    plt.grid(True)
    plt.show()


    print("\nSystem Information:")
    print(f"CPU: {get_cpu_info()['brand_raw']}")
    print(f"RAM: {psutil.virtual_memory().total / 1e9:.2f} GB")
    print(f"Python Version: {platform.python_version()}")


benchmark_sorting_algorithms()
```
